```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the mtcars dataset
url = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
mtcars = pd.read_csv(url)
```

```python
# Extract independent (hp) and dependent (mpg) variables
X = mtcars['hp'].values  # Independent variable
y = mtcars['mpg'].values  # Dependent variable
```

```python
# Transform the data for polynomial regression (e.g., degree 2)
degree = 2
X_poly = np.column_stack([X**i for i in range(degree + 1)])  # Add powers of X up to degree
```

```python
# Compute the polynomial regression coefficients
coefficients = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y
```

```python
# Generate predictions
y_pred = X_poly @ coefficients
```

```python
# Visualize the results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(np.sort(X), y_pred[np.argsort(X)], color='red', label='Polynomial Fit')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles per Gallon (mpg)')
plt.title(f'Polynomial Regression (Degree {degree})')
plt.legend()
plt.show()
```

```python
# Display the equation
terms = [f"{coefficients[i]:.2f} * X^{i}" for i in range(degree + 1)]
equation = " + ".join(terms).replace("* X^0", "").replace("X^1", "X")
print(f"Polynomial Regression Equation: mpg = {equation}")
```
