```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = load_iris()
X = iris.data
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
centroids = kmeans.cluster_centers_
```

```python
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df['Cluster'] = labels
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=iris_df['sepal length (cm)'], y=iris_df['sepal width (cm)'], hue=iris_df['Cluster'], palette='Set1', s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='yellow', marker='X', label='Centroids')  # Centroids
plt.title("K-Means Clustering of Iris Data (Sepal Length vs Sepal Width)")
plt.legend()
plt.show()
```
